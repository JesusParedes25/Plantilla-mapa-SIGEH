```python
import folium
import pandas as pd
import geopandas as gpd
import json
from folium.plugins import MeasureControl

# Crea un mapa base con un centro y un nivel de zoom determinados
m = folium.Map(location=[20.11, -98.73], zoom_start=9,tiles="cartodbpositron")
folium.TileLayer(
    tiles='https://{s}.tile.openstreetmap.org/{z}/{x}/{y}.png',
    attr='&copy; <a href="https://openstreetmap.org/copyright">OpenStreetMap</a> contributors',
    name='Open Street Map'
).add_to(m)

folium.TileLayer(
    tiles='cartodbdark_matter',
    name='Obscuro'
).add_to(m)

#Agrego capas 
gdf = gpd.read_file("delitos.geojson")
pueblos=gpd.read_file("Pueblos historicos.geojson")

#Creación de capas para Layer control
municipios= folium.FeatureGroup(name="Municipios")
pueblos_indigenas= folium.FeatureGroup(name="Pueblos indígenas")

#Agrego el geojson al mapa
folium.GeoJson(gdf)
folium.GeoJson(pueblos).add_to
datos = pd.read_csv("delitos.csv")
datos_pueblos = pd.read_csv("Pueblos historicos.csv")

#Creo un estilo 
bordersStyle={'color':'brown',
              'weight':1,
              'fillColor':'none',
              'line_opacity':0.5,
             }
#Asigno el color a la capa municipios-(gdf)
folium.GeoJson(gdf,
               name='Municipios',
               style_function=lambda x:bordersStyle).add_to(m)


#Colorear iconos de puntos de acuerdo a un campo de pueblos indigenas
for i in range(0,len(datos_pueblos)):
    html=f"""
        <h1> {'Municipio: '+datos_pueblos.iloc[i]['NOM_MUN']+' '+' <br><br>Pueblo histórico: '+datos_pueblos.iloc[i]['pueblo_ah']}</h1>
        <p>You can use any html here! Let's do a list:</p>
        <ul>
            <li>Item 1</li>
            <li>Item 2</li>
        </ul>
        </p>
        <p>And that's a <a href="https://inabisa.com.mx/pdf/EC6">link</a></p>
        """
    iframe = folium.IFrame(html=html, width=400, height=220)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
      location=[datos_pueblos.iloc[i]['y'], datos_pueblos.iloc[i]['x']],
       icon=folium.Icon(color=datos_pueblos.iloc[i]['colors'],prefix='fa',icon='school'),
      popup=popup,
    ).add_to(pueblos_indigenas)
 #agrego la capa coloreada a m   
pueblos_indigenas.add_to(m)

#Colorear en coropleta la capa de municipios de acuerdo a un campo
"""pueb=folium.Choropleth(
    geo_data=gdf,
    data=datos,
    columns=["id", "VIO_FAM"],
    key_on="feature.properties.id",
    fill_color="RdYlGn_r",
    fill_opacity=0,
    line_opacity=1,
    line_color='black',
    nan_fill_color="white",
    legend_name="Llamadas por cada 100 mil habitantes",
)"""


folium.LayerControl(collapsed=False, position="topleft", title="Custom Title").add_to(m)
m.add_child(MeasureControl())


m
m.save('map.html')
```
